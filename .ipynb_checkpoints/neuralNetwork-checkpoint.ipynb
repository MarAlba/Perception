```python
%pylab inline
import cv2
import glob
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
import scipy
import scipy.io
import matplotlib.pyplot as plt
import csv
```

```python
from keras.applications.vgg19 import decode_predictions
import csv
```

```python
model = tf.keras.applications.VGG19(
        include_top=True,
        weights="imagenet",
        input_tensor=None,
        input_shape=None,
        pooling=None,
        classes=1000,
    )
```

```python
image = cv2.imread('../Data/nnData/0003.png')
resize_number = min(224/image.shape[1],224/image.shape[0])
resized_image = cv2.resize(image, (0,0), fx=resize_number, fy=resize_number) 
zeropad_image = cv2.copyMakeBorder( resized_image,int(ceil((224-resized_image.shape[0])/2)), int(floor((224-resized_image.shape[0])/2)), int(ceil((224-resized_image.shape[1])/2)), int(floor((224-resized_image.shape[1])/2)), cv2.BORDER_REPLICATE)
plt.imshow(zeropad_image)
yhat = model.predict(np.array([zeropad_image]))
# convert the probabilities to class labels
label = decode_predictions(yhat)
# retrieve the most likely result, e.g. highest probability
label = label[0][0]
# print the classification
print(decode_predictions(yhat))
print('%s (%.2f%%)' % (label[1], label[2]*100))
out = decode_predictions(yhat, top = 250)
itemlist = ["n02840245","n07930864","n02971356"]

outfile=open('./immates.csv','w')
writer=csv.writer(outfile)
#writer.writerow(["SNo", "States", "Dist", "Population"])
writer.writerows(itemlist)

with open('./immates.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            
            line_count += 1
    print(f'Processed {line_count} lines.')

itemarray = np.array([])
for index, item in enumerate(out[0]):
    for i in itemlist:
        if (item[0]) == (i):
            itemarray=np.append(itemarray,item)
            print ( str(index) + " " + str(item))
reshaped_itemarray = itemarray.reshape(-1,3)
print(reshaped_itemarray[argmax(reshaped_itemarray[:,2])][1])
```
